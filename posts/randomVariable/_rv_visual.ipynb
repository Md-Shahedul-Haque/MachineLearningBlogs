```python
#| label: pmf-binomial
#| code-fold: false
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import binom
# parameters for the binomial distribution
n = 10
p = 0.5

# generate the values for x
x = np.arange(0, n+1)

# calculate the PMF
pmf = binom.pmf(x, n, p)
```

```python
#| label: fig-pmf
#| fig-cap: "Probability Mass Function (PMF) of a discrete random variable with binomial distribution"
# create the plot
plt.bar(x, pmf)
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.title('PMF of a Binomial Distribution (n=10, p=0.5)')
plt.show()
```

## PDF

```python
#| label: pdf-normal
#| code-fold: false
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

# parameters for the normal distribution
mu = 0
sigma = 1

# generate the values for x
x = np.linspace(-10, 10, 1000)
# calculate the PDF
pdf = norm.pdf(x, mu, sigma)
```

```python
#| label: fig-pdf
#| fig-cap: "Probability Density Function (PDF) of a continous random variable with normal distribution"

# create the plot
plt.plot(x, pdf)
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('PDF of a Normal Distribution (mu=0, sigma=1)')
plt.show()
```

## CDF

```python
#| label: cdf-normal
#| code-fold: false
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

# parameters for the normal distribution
mu = 0
sigma = 1

# generate the values for x
x = np.linspace(-10, 10, 1000)
# calculate the PDF
cdf = norm.cdf(x, mu, sigma)
```

```python
#| label: fig-cdf
#| fig-cap: "Cumulative Distribution Function (CDF) of a continous random variable with normal distribution"


# create the plot
plt.plot(x, cdf)
plt.xlabel('x')
plt.ylabel('Cumulative Probability')
plt.title('CDF of a Normal Distribution (mu=0, sigma=1)')
plt.show()
```
